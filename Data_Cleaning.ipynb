```python
# Convert 'B_BIRTHDATE' and any date-related columns to datetime
data_cleaned2['B_BIRTHDATE'] = pd.to_datetime(data_cleaned2['B_BIRTHDATE'], errors='coerce')

# Convert numerical columns to float/int
numeric_columns = ['VALUE_ORDER', 'AMOUNT_ORDER']  # Adjust this list as necessary
data_cleaned2[numeric_columns] = data_cleaned[numeric_columns].apply(pd.to_numeric, errors='coerce')

# Check data types
data_cleaned.dtypes
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Assuming your dataset is stored in a CSV file
# df = pd.read_csv('your_dataset.csv')

# For this example, I'll assume your DataFrame is already loaded
# Replace 'VALUE_ORDER_PRE' with the actual column name if needed
# df['VALUE_ORDER_PRE']

# Plot the distribution of non-zero values
non_zero_values = data_cleaned2[data_cleaned2['VALUE_ORDER_PRE'] != 0.0]['VALUE_ORDER_PRE']

plt.figure(figsize=(10, 6))
sns.histplot(non_zero_values, bins=50, kde=True)  # kde=True adds a smooth line for distribution

plt.title('Distribution of Non-Zero Values in VALUE_ORDER_PRE', fontsize=16)
plt.xlabel('VALUE_ORDER_PRE', fontsize=12)
plt.ylabel('Frequency', fontsize=12)
plt.grid(True)

# Show the plot
plt.show()
```

```python
data_cleaned2['VALUE_ORDER_PRE'] = data_cleaned2['VALUE_ORDER_PRE'].fillna(data_cleaned2['VALUE_ORDER_PRE'].median())
```

```python
data_cleaned2.head(20)
```

```python
zero_value_count = (data_cleaned2['VALUE_ORDER_PRE'] == 0.0).sum()
print(f"Number of rows with 0.0 in VALUE_ORDER_PRE: {zero_value_count}")
```

```python
median_value = data_cleaned2[data_cleaned2['VALUE_ORDER_PRE'] != 0.0]['VALUE_ORDER_PRE'].median()
print(f"Median of non-zero values in VALUE_ORDER_PRE: {median_value}")

# Replace the 0.0 values with the calculated median
data_cleaned2['VALUE_ORDER_PRE'] = data_cleaned2['VALUE_ORDER_PRE'].replace(0.0, median_value)

# Verify that 0.0 values have been replaced
new_zero_value_count = (data_cleaned2['VALUE_ORDER_PRE'] == 0.0).sum()
print(f"Number of rows with 0.0 after replacement: {new_zero_value_count}")
```
